```python
# Q.3 Explain the following terms:
      # • The difference in np.array, np.asarray and np.asanyarray.
      # • The difference between Deep copy and shallow copy.

## Ans:

# The difference in np.array, np.asarray and np.asanyarray:
     # np.array:
              # Purpose: Creates a new array from any object exposing the array interface.
              # Behavior: Always creates a new array, even if the input is already an array.
              # Use Case: Use np.array when you need a new array and want to ensure that any modifications to the new array do not affect the original.
              # Example:
                       import numpy as np
                       a = [1, 2, 3]
                       b = np.array(a)
                       print(b)     # Output: [1 2 3]
      # np.asarray:
                 # Purpose: Converts the input to an array, but does not create a copy if the input is already an array.
                 # Behavior: If the input is already an array, it returns the input array itself.
                 # Use Case: Use np.asarray when you want to avoid unnecessary copying, especially for existing arrays.
                 # Example:
                          import numpy as np
                          a = np.array([1, 2, 3])
                          b = np.asarray(a)
                          print(b is a)     # Output: True
      # np.asanyarray:
                    # Purpose: Similar to np.asarray, but passes through subclasses of ndarray.
                    # Behavior: If the input is a subclass of ndarray, it returns the input itself.
                    # Use Case: Use np.asanyarray when you want to retain subclass behavior in the output.
                    # Example:
                             import numpy as np
                             class MyArray(np.ndarray):
                                 pass

                             a = MyArray((2, 2))
                             b = np.asanyarray(a)
                             print(b is a)     # Output: True

# The difference between Deep copy and shallow copy:
      # Deep Copy:
                # Definition: Creates a new object and recursively copies all objects found in the original.
                # Behavior: Changes to objects in the copied object do not affect the original object.
                # Use Case: Use a deep copy when you need a completely independent copy of the object and its nested structures.
                # Example:
                         import numpy as np

                         original_array = np.array([[1, 2, 3], [4, 5, 6]])
                         deep_copy = original_array.copy()
                         deep_copy[0][0] = 99
                         print(original_array)     # Output: [[1 2 3]
                                                               [4 5 6]]

      # Shallow Copy:
                   # Definition: Creates a new object, but inserts references into it to the objects found in the original.
                   # Behavior: Changes to mutable objects in the copied object will reflect in the original object.
                   # Use Case: Use a shallow copy when you want to create a new object that shares references to the same inner objects, and you’re okay with modifications affecting the original.
                   # Example:
                           import numpy as np

                           original = np.array([[1, 2, 3], [4, 5, 6]])
                           shallow_copy = original[:]
                           shallow_copy[0][0] = 99
                           print(original)      # Output: [[99  2  3]
                                                           [ 4  5  6]]
```

```python
# Q.5 Create a NumPy array with random integers between 1 and 10 of shape (5, 6). After creating the array perform the following operations:
      # a) Extract all even integers from array.
      # b) Extract all odd integers from array.

## Ans:

import numpy as np

array = np.random.randint(1, 11, (5, 6))    # Create a 5x6 array with random integers between 1 and 10
print("Random Array:\n", array)

# Extract all even integers
even_integers = array[array % 2 == 0]
print("\nEven integers:\n", even_integers)

# Extract all odd integers
odd_integers = array[array % 2 != 0]
print("\nOdd integers:\n", odd_integers)
```

```python
# Q.6 Create a 3D NumPy array of shape (3, 3, 3) containing random integers between 1 and 10. Perform the following operations:
      # a) Find the indices of the maximum values along each depth level (third axis).
      # b) Perform element-wise multiplication of between both array.

## Ans:

import numpy as np

# Create a 3D NumPy array of shape (3, 3, 3) containing random integers between 1 and 10
array = np.random.randint(1, 11, (3, 3, 3))
print("Original 3D array:\n", array)

# Find the indices of the maximum values along each depth level (third axis)
max_indices = np.argmax(array, 2)
print("\nIndices of maximum values along each depth level:\n", max_indices)

# Perform element-wise multiplication of the array with itself
multiplied_array = array * array
print("\nElement-wise multiplied array:\n", multiplied_array)
```

```python
# Q.14 Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and Perform the following tasks using Matplotlib:
       # a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and right y-axis for 'Humidity').
       # b) Label the x-axis as 'Date'.
       # c) Set the title of the plot as 'Temperature and Humidity Over Time'.

## Ans:

import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Date': pd.date_range(start='2021-01-01', end='2021-12-31', freq='D'),
    'Temperature': np.random.randint(10, 40, 365),
    'Humidity': np.random.randint(30, 90, 365)
}

df = pd.DataFrame(data)

# a) Create the plot
fig, ax1 = plt.subplots()

ax1.plot(df['Date'], df['Temperature'], color='blue', label='Temperature')    # Plot Temperature on the left y-axis
ax1.set_ylabel('Temperature (Celsius)', color='blue')

ax2 = ax1.twinx()    # Create a second axis for Humidity on the right y-axis

ax2.plot(df['Date'], df['Humidity'], color='red', label='Humidity')    # Plot Humidity on the right y-axis
ax2.set_ylabel('Humidity (%)', color='red')

# b) Set the x-axis label
plt.xlabel('Date')
ax1.set_xlabel('Date', color='black')

# c) Set the title of the plot
plt.title('Temperature and Humidity Over Time')
plt.show()
```

```python
# Q.15 Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following tasks using Matplotlib:
       # a) Plot a histogram of the data with 30 bins.
       # b) Overlay a line plot representing the normal distribution's probability density function (PDF).
       # c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.
       # d) Set the title of the plot as 'Histogram with PDF Overlay'.

## Ans:

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

data = np.random.normal(0, 1, 1000)    # Generate 1000 samples from a normal distribution

# a) Plot the histogram of the data with 30 bins
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Histogram')

# b) Overlay the PDF of the normal distribution
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, data.mean(), data.std())
plt.plot(x, p, linewidth=2,color='red', label='PDF')

# c) Label the x-axis and y-axis
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

# d) Set the title of the plot
plt.title('Histogram with PDF Overlay')

plt.legend()
plt.show()
```

```python
# Q.17 Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

## Ans:

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(0)
x = np.random.randn(100)
y = np.random.randn(100)

quadrants = np.where(x > 0, np.where(y > 0, 'Quadrant I', 'Quadrant IV'),
                     np.where(y > 0, 'Quadrant II', 'Quadrant III'))

# Create scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x=x, y=y, hue=quadrants, palette='deep', s=100)

# Add legend, labels, and title
plt.legend(title='Quadrant')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.show()
```

```python
# Q.18 With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'.

## Ans:

import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()

x = np.linspace(0, 2 * np.pi, 100)  # X values from 0 to 2π
y = np.sin(x)  # Sine of X values

p = figure(title='Sine Wave Function', x_axis_label='X', y_axis_label='sin(X)',
           width=800, height=400)

p.grid.grid_line_alpha = 0.3
p.line(x, y, line_width=2, color='blue', legend_label='sin(x)')
show(p)
```

```python
# Q.19 Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'.

## Ans:

import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.transform import factor_cmap
from bokeh.palettes import Category20

output_notebook()

categories = ['A', 'B', 'C', 'D', 'E', 'F']
values = np.random.randint(1, 100, size=len(categories))

data = pd.DataFrame({'categories': categories, 'values': values})
source = ColumnDataSource(data)

p = figure(x_range=categories, title="Random Categorical Bar Chart",
           toolbar_location=None, tools="")

p.vbar(x='categories', top='values', width=0.9, source=source, legend_field="categories",
       line_color='white', fill_color=factor_cmap('categories', palette=Category20[len(categories)], factors=categories))

hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)
p.xaxis.axis_label = "Categories"
p.yaxis.axis_label = "Values"
p.y_range.start = 0
p.xgrid.grid_line_color = None
p.legend.orientation = "horizontal"
p.legend.location = "top_center"
show(p)
```
